### Passo 1 -  Importar os dados

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Importar a planilha

candy_data = pd.read_csv("candy.csv", index_col="id")
candy_data
```

```python
# Printar as últimas 5 linhas

candy_data.head()
```

O conjunto de dados contém 83 linhas, onde cada uma corresponde a uma barra de chocolate diferente. São 13 colunas:

-> 'competitorname'- contém o nome das barras de chocolate.

As próximas 9 colunas (de 'chocolate' a 'pluribus') descrevem cada doce. Por exemplo, linhas com bombons de chocolate têm "Sim" na coluna 'chocolate', e bombons sem chocolate têm "Não" na mesma coluna).

-> 'sugarpercent'- fornece alguma indicação da quantidade de açúcar, onde valores mais altos significam maior teor de açúcar.

-> 'pricepercent'- mostra o preço por unidade, em relação aos outros doces no conjunto de dados.

-> 'winpercent' - é calculado a partir dos resultados da pesquisa; valores mais altos indicam que o doce era mais popular entre os entrevistados

### Passo 2 -  Perguntas

#### Pergunta 1: Qual doce foi mais popular entre os entrevistados da pesquisa: '3 Mosqueteiros' ou 'Almond Joy'?

```python
#3 Mosquereiros

doce_popular1 = candy_data.head().iloc[[1]].max()  

print(doce_popular1['winpercent'])
```

```python
# Almond Joy

doce_popular2 = candy_data.iloc[[3]].max()  
print(doce_popular2['winpercent'])
```

O doce 3 Mosquereiros foi o mais popular entre os entrevistados da pesquisa com 67,60, comparado ao Almond Joy.

#### Pergunta 2: Qual doce tem o maior teor de açúcar: 'Air Heads' ou 'Baby Ruth'?

```python
# Air Heads

teor_açucar1 = candy_data.head().iloc[[2]].max()

print(teor_açucar1['sugarpercent'])
```

```python
# Baby Ruth

teor_açucar2 = candy_data.head().iloc[[4]].max()

print(teor_açucar2['sugarpercent'])
```

O doce Air Heads possui o maior teor de açúcar comparado ao Baby Ruth, apresentando o valor de 0.90

### Passo 3 - Plotar o gráfico de dispersão

Crie um gráfico de dispersão que mostre a relação entre 'sugarpercent' (no eixo x horizontal) e 'winpercent' (no eixo y vertical).

```python
# scatterplot - gráfico de dispersão

sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

### Passo 4 - Plotar gráfico de regressão

Crie o mesmo gráfico de dispersão criado na **Etapa anterior**, porém, com uma linha de regressão

```python
# regplot - linha de regressão

sns.regplot(x=candy_data['sugarpercent'],y=candy_data['winpercent'])
```

**Pergunta:**De acordo com o gráfico acima, existe uma ligeira correlação entre 'winpercent' e 'sugarpercent'? O que isso lhe diz sobre o doce que as pessoas tendem a preferir?


**Resposta:**Como a linha de regressão tem uma inclinação ligeiramente positiva, isso nos diz que há uma correlação ligeiramente positiva entre 'winpercent' e 'sugarpercent'. Assim, as pessoas têm uma leve preferência por doces contendo relativamente mais açúcar

### Passo 5 - Chocolate

Crie um gráfico de dispersão para mostrar a relação entre 'pricepercent' (no eixo x horizontal) e 'winpercent' (no eixo y vertical). Use a coluna 'chocolate' para codificar os pontos com cores.

```python
# Gráfico de dispersão

sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
```

### Passo 6 - Investigando o Chocolate


Crie o mesmo gráfico de dispersão criado na etapa anterior, mas agora com duas linhas de regressão, correspondendo a (1) bombons de chocolate e (2) bombons sem chocolate.

Use sns.lmplot e defina as variáveis ​​para o eixo x, eixo y, cor dos pontos e o conjunto de dados.

```python
# Lmplot - duas linhas de regressão

sns.lmplot(x="pricepercent", y="winpercent", hue="chocolate", data=candy_data)
```

**Pergunta:** Usando as linhas de regressão, o que é possível conlcuir sobre os efeitos do chocolate e do preço na popularidade dos doces?

**Resposta:** Em relação à linha de regressão para bombons de chocolate. Como essa reta tem uma inclinação levemente positiva, podemos dizer que bombons de chocolate mais caros tendem a ser mais populares (do que bombons de chocolate relativamente mais baratos). Da mesma forma, como a linha de regressão para doces sem chocolate tem inclinação negativa, podemos dizer que se os doces não contêm chocolate, eles tendem a ser mais populares quando são mais baratos. Uma observação importante, no entanto, é que o conjunto de dados é bem pequeno -- portanto, não devemos confiar demais nesses padrões! Para inspirar mais confiança nos resultados, devemos adicionar mais doces ao conjunto de dados.

### Passo 7 -  Todo mundo ama Chocolate

Crie um gráfico de dispersão categórico para destacar a relação entre 'chocolate' e 'winpercent'. Coloque 'chocolate' no eixo x (horizontal) e 'winpercent' no eixo y (vertical).

Use sns.swarmplot e defina as variáveis para os eixos x e y.
