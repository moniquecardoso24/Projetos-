### Importar bibliotecas

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Ler o arquivo

```python
ign_data = pd.read_csv("ign_scores.csv", index_col="Platform")
ign_data
```

### Responder as perguntas

O conjunto de dados impresso mostra a pontuação média, por plataforma e gênero. Use os dados para responder as questões abaixo.

1.Qual é a maior pontuação média recebida pelos jogos de **PC**, para qualquer gênero?

2.Na plataforma **Playstation Vita**, qual gênero tem a menor média de pontuação?


```python
# Pergunta 1 - O maior pontuação recebida por PC games

valor_max = ign_data.loc[["PC"]].max(axis = 1) # axis=1 valor max da linha

print(valor_max)
```

```python
# Pergunta 2 - Menor pontuação da PlayStation Vita

valor_min = ign_data.loc[["PlayStation Vita"]].min(axis = 1)  # axis=1 valor min da linha

print(valor_min)
```

### Plotar o gráfico de barras

#### Qual plataforma é a melhor?

Crie um gráfico de barras que mostre a pontuação média para jogos de **corrida**, para cada plataforma. O gráfico deve ter uma barra para cada plataforma.

```python
# Gráfico de Barras

# Definir a largura e a altura da figura
plt.figure(figsize=(8,6))

# Gráfico de barras mostrando pontuação média para jogos de corrida por plataforma

sns.barplot(x=ign_data["Racing"], y=ign_data.index)  # coluna racing

# Adicionar rótulo para o eixo horizontal
plt.xlabel("Classificação de 0 a 8")

# Adicionar rótulo para o eixo vertical
plt.title("Average Score for Racing Games, by Platform")
```

**Pergunta:** Com base no gráfico de barras, é possível esperar que um jogo de corrida para a plataforma Wii receba uma classificação alta? Se não, qual plataforma de jogo parece ser a melhor alternativa?


**Resposta:** Com base nos dados, não se pode esperar que um jogo de corrida para a plataforma Wii receba uma classificação alta. De acordo com o gráfico, em média, jogos de corrida para Wii pontuam mais baixo(5 pontos) do que qualquer outra plataforma. O Xbox One parece ser a melhor alternativa, já que possui as avaliações médias mais altas(acima de 8 pontos).


### Plotar o Mapa de calor

Use os dados para criar um mapa de calor de pontuação média por gênero e plataforma.

```python
# Mapa de Calor - heatmap

# Definir a largura e a altura da figura
plt.figure(figsize=(11,11))

# Mapa de calor mostrando a pontuação média do jogo por plataforma e gênero

sns.heatmap(ign_data, annot=True)  

# Adicionar rótulo para o eixo horizontal
plt.xlabel("Genre")

# Adicionar rótulo para o eixo vertical
plt.title("Average Game Score, by Platform and Genre")
```

**Pergunta:** Qual combinação de gênero e plataforma recebe as classificações médias mais altas? Qual combinação recebe as classificações médias mais baixas?

**Resposta:** De acordo com o Mapa de calor, os jogos de Simulation(Simulação) para Playstation 4 recebem as avaliações médias mais altas (9,2). Jogos de Shooting(tiro) e Fighting(luta) para a coluna "Game Boy Color" recebem as classificações médias mais baixas (4,5)
